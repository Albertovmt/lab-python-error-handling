def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
def manage_orders():
    def initialize_inventory(products):
        inventory = {}
        for product in products:
            while True:
                try:
                    quantity = int(input(f"Enter quantity of {product}: ").strip())
                    if quantity >= 0:
                        quantity = int(quantity)
                        inventory[product] = quantity
                        break
                    else: 
                        print("Quantity cannot be negative. Please enter a valid quantity")
                except ValueError:
                        print('Invalid input. Please enter a valid quantity')

        print("------Availability of products------")
        for item,value in inventory.items():
            print(item,":",value)  
                
        return inventory

    def get_customer_orders(inventory):
        customer_orders = set()

        while True:
            try:
                order = input('Enter the name of the product to purchase: ').strip().lower()

                if order not in inventory:
                    raise KeyError
                if inventory[order] <= 0:
                    raise ValueError
            
                customer_orders.add(order)
                inventory[order] -= 1
            except KeyError:
                 print(f'Product {order} not found')
            except ValueError:
                 print(f'Sorry, {order} is out of stock')
        
            validation = input("Do you want to buy more items? (yes/no): ").strip().lower()
            while validation not in ('yes','no'):
                print(f"'{validation}' is not a correct option. Please select 'yes' or 'no'.", "\n")
                validation = input("Do you want to buy more items? (yes/no): ").strip().lower() 

            if validation == "no":
                break 

        print("------Customer´s order------")
        print (f"The customer has ordered {customer_orders}", "\n")

        return customer_orders
    
    def calculate_order_statistics(customer_orders,products):

        total_products_ordered = len(customer_orders)
        percentage_ordered = round((total_products_ordered/len(products))*100,1)

        return total_products_ordered, percentage_ordered

    def print_order_statistics(order_statistics):

        total_products_ordered, percentage_of_products_ordered = order_statistics
        print("------Order Statistics------")
        print(f"Total products ordered: {total_products_ordered}")
        print(f"Percentage of products ordered: {percentage_of_products_ordered}%","\n")

    def print_updated_inventory(inventory):
        print("------Updated inventory------")
        for key,value in inventory.items():
            print(f"{key}: {value}")
        print()

    def total_price(customer_orders):

        prices_list = []
        for product in customer_orders:
            while True:
                try:
                    user_input = float(input(f'Enter the price of one {product}').replace(',','.'))
                    if user_input >= 0:
                        prices_list.append(user_input)
                        break
                    else:
                        print('Price cannot be negative. Please enter a valid price')
                except ValueError:
                    print('Invalid input. Price must be a number.')

        print('------Receipt------')
        for price,item in zip(prices_list,customer_orders):
            print(f'The price of one {item} is: {price}$ ') 

        sum_prices = round(sum(prices_list),1)
        
        return sum_prices

    products = ["t-shirt", "mug", "hat", "book", "keychain"]

    inventory = initialize_inventory(products) 
    customer_orders = get_customer_orders(inventory)
    order_statistics = calculate_order_statistics(customer_orders,products)
    print_order_statistics(order_statistics)
    print_updated_inventory(inventory)
    total = total_price(customer_orders)
    print(f'the total price is: {total}$')
````
